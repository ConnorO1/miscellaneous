```python
from matplotlib.image import imread # you could also use `imageio.imread()`
image = imread("ladybug.jpg")
image.shape
```

```python
X = image.reshape(-1, 3)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8).fit(X)
```

```python
import numpy as np
test = np.array([1, 1, 1, 2, 2, 2])
```

```python
# we can return the cluster centre (mean) for each pixel value -> 
# put array as index into another array, we will return that sequence of array elements

segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(segmented_img/255)
```
