### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()

X, y = make_moons(n_samples = 1000, noise = 0.05)
```

```python
sns.scatterplot(x = X[:, 0], y = X[:, 1], hue = y, alpha = 0.2)
```

```python
db = DBSCAN(eps = 0.2)
db.fit(X)
palette = sns.color_palette("hls", 2)
plt.figure(figsize = (8, 8))
sns.scatterplot(x = X[:, 0], y = X[:, 1], hue = db.labels_, alpha = 0.2, palette = palette)
```

```python
db.components_
```

```python
db.labels_
```

```python
# predict cluster sampls belong to

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)

# train on core_samples from moons

knn.fit(db.components_, db.labels_[db.core_sample_indices_])
knn.predict_proba(np.array([[1, 0.25],[2,0],[0.5, 0.25]]))
```
